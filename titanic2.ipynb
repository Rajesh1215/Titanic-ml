```python
import numpy as np
import matplotlib.pyplot as plt

# Set the mean and standard deviation for the normal distribution
mean = 0
std_dev = 1

# Generate 100 random samples following a normal distribution
random_samples = np.random.normal(loc=mean, scale=std_dev, size=100)

# Calculate the Z-scores for each data point
z_scores = (random_samples - np.mean(random_samples)) / np.std(random_samples)

# Create a bar plot of Z-scores on the x-axis and values on the y-axis
plt.bar(z_scores, random_samples, edgecolor='black', alpha=0.75,width=0.1)

# Add labels and title
plt.xlabel('Z-Score')
plt.ylabel('Values')
plt.title('Bar Plot of Z-Scores for Normal Array')

# Show the bar plot
plt.show()
```

```python
class MyClass:
    def __init__(self):
        self._radius = 0

    def radius(self, new_radius):
        # Validation and other custom logic can be performed here
        if new_radius >= 0:
            self._radius = new_radius
        else:
            raise ValueError("Radius must be a non-negative value.")

    @property
    def radius(self):
        return self._radius

    def hi(self):
        print("hi")


# Create an instance of the class
obj = MyClass()

# Call the method to set the radius
obj.set_radius=5

# Access the updated radius using the property
print(obj.radius)  # Output: 5

# Try to set an invalid negative radius value
try:
    obj.set_radius(-1)
except ValueError as e:
    print("Error:", e)  # Output: Error: Radius must be a non-negative value.

# Access the radius property after the exception
print(obj.radius)  # Output: 5

# Call the 'hi' method
obj.hi()  # Output: hi
```
